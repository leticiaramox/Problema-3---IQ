# Desigualdades de Bell: Do Paradoxo EPR à Geometria da Não-Localidade

## 1. Geometria local vs. quântica¶

Segundo Bluhm et al. (2023)[1], um polítopo $P$ contendo a origem pode ser descrito de duas formas equivalentes:

1. por suas faces, como produto da interseção de meio-espaços (a representação H):
$$
P = \bigcap_{i=1}^{f} \{ x \in \mathbb{R}^g : \langle x, h_i \rangle \le 1 \}.
$$
2. por seus pontos extremos, como um casco convexo (a representação V):
$$
P = \operatorname{conv}\{ v_i \}_{i=1}^k
$$

Ou seja, para a representação V, por exemplo, um politopo é o fecho convexo de um número **finito** de pontos extremos. Sabendo que cada ponto extremo é uma estratégia determinística (Alice sempre responde $a = f(x)$ e Bob $b = g(y)$), existem finitas funções determinísticas possíveis. Pontanto, o conjunto local é um politopo, porque é o casco convexo de um conjunto finito de pontos determinísticos.

Por outro lado, um conjunto quântico não precisa ser um politopo, pois pontos extremos quânticos não são finitos. As responsáveis por descrever os estados quânticos são as matrizes de densidade $\mathcal{M}_{1,+}^d(\mathbb{C})$, o que as leva a ter infinitas dimensões e parâmetros contínuos. Por isso, não existe lista finita de desigualdades que caracterize o conjunto quântico (diferente do local). 

Nesse conceito, é possível observar como a curva quântica "fura" a face CHSH no gráfico abaixo: 

```python
from PIL import Image
import matplotlib.pyplot as plt

# abrir a imagem
img = Image.open("Figure_3_3.png")

# mostrar a imagem
plt.imshow(img)
plt.axis("off")
plt.show()
```

Na figura acima, o polítopo local é representado através de um tetraedro, devido à representação do quatro quibits, e a curva quântica tem formato ondulatório. Nela, uma parte da curva está contida dentro do politopo local e outra está por fora. Perceba, também, que as extremidades da onda estão posicionadas exatamente nos vértices da figura em azul (o tetraedro). Segundo Augusiak et al. (2014), as correlações locais formam um politopo (região azul-acinzentada), cujos vértices são dados pelas correlações locais determinísticas. Esse conjunto constitui um subconjunto próprio do conjunto das correlações quânticas. Correlações que não pertencem ao polítopo são denominadas não locais (parte da curva que excede a região do tetraedro) e podem ser sempre detectadas por meio de uma desigualdade de Bell.

## 2. Mapa de decisão

Os diferentes valores de S podem ser estudados a partir da desigualdade de Bell. 

#### 1. $S = 2$
As medições realizadas tanto por Alice quanto por Bob têm apenas dois resultados possíveis: ±1. Diz-se que elas são medições dicotômicas, isto é, medições, por exemplo, do tipo sim/não, para cima/para baixo, etc. As medições ($A$, $A'$) e ($B$, $B'$) são repetidas $n$ vezes. Seja ($a_j$, $a'_j$, $b_j$, $b'_j$) o conjunto de resultados da j-ésima medição.

Com isso, sabe-se que 
$$
\left| (a_j + a'_j)\, b_j + (a_j - a'_j)\, b'_j \right| \le 2,\ \forall j.
\tag{1}
$$
dado que $|b_j| = |b'_j| = 1$. Além disso, é válido lembrar que o valor máximo de $|a_j + a'_j|$  é igual a $2$, ou seja, $a_j = a'_j = \pm 1$. Contudo, quando $|a_j + a'_j| = 2$, pode-se observar que $|a_j - a'_j|$ vai a 0, mostrando, assim, a validade inequação acima. Logo:
$$
\left| \left\langle (A + A')\,B + (A - A')\,B' \right\rangle_{\mathrm{av}} \right| \le 2
\tag{2}
$$
Ademais, por definição, temos que
$$
S = \langle AB \rangle + \langle AB' \rangle + \langle A'B \rangle - \langle A'B' \rangle
\tag{3}
$$
ou seja
$$
\left| \left\langle (A + A')\,B + (A - A')\,B' \right\rangle_{\mathrm{av}} \right| \le S \le 2
\tag{4}
$$
Nesse conceito, conclui-se que, para $S = 2$ (valor máximo), as correlações podem ser explicadas por um modelo clássico local, não havendo prova de não-localidade (note que em momento nenhum foi introduzido conceitos relacionados à quântica do problema).

#### 2. $S=2\sqrt{2}$

Quando fenômenos quânticos são contabilizados, a equação (1) é modificada e se torna a equação (5):
$$
\langle \psi | \mathcal{C} | \psi \rangle,
\tag{5}
$$
em que $|\psi\rangle$ representa o estado normalizado do sistema, $\langle \psi | \psi \rangle = 1$, e $\mathcal{C}$ é dado pelo operador  Hermitian Bell--CHSH:
$$
\mathcal{C} = (A + A') \otimes B + (A - A') \otimes B',
\tag{6}
$$
A pergunta que nos resta é se a equação (2) é válida a nível quâtico, ou seja:
$$
\left| \langle \psi | \mathcal{C} | \psi \rangle \right| \le 2 \, ?.
\tag{7}
$$
A resposta obtida é surpreendente. Se o estado $|\psi\rangle$ é um **estado produto**, a relação (1) ainda é válida. Porém, se o estado $|\psi\rangle$  possui emaranhamento, então a relação (4) é estritamente maior do que $2$. Em outras palavras:
$$
\left| \langle \psi | \mathcal{C} | \psi \rangle \right|_{\mathrm{emaranhado}} > 2.
\tag{8}
$$
Nesse cenário, encontra-se uma violação da inequação de Bell--CHSH. Assim como feito em $[4]$, como primeiro exemplo de violação da desigualdade de Bell--CHSH, consideramos o caso em que Alice e Bob compartilham o primeiro estado emaranhado de Bell $|\phi_0\rangle$. Assim, afirma-se que
$$
\langle \phi_0 | \mathcal{C} | \phi_0 \rangle 
= \langle \phi_0 | (A + A') \otimes B + (A - A') \otimes B' | \phi_0 \rangle 
= \cos(\alpha + \beta) + \cos(\alpha' + \beta) + \cos(\alpha + \beta') - \cos(\alpha' + \beta')
\tag{9}
$$
na qual $(\alpha, \alpha', \beta, \beta')$ são parâmetros arbitrários. A escolha dos ângulos $\alpha = 0$, $\alpha' = \pi/2$, $\beta = -\pi/4$ e $\beta' = \pi/4$ é padrão na literatura porque maximiza a violação da desigualdade de Bell--CHSH para o estado de Bell $|\phi_0\rangle$. Esses valores correspondem a direções de medição mutuamente desfasadas por $45^\circ$, o que otimiza as correlações quânticas previstas pela mecânica quântica. 

Fazendo as mesma escolhas de valores que [5] para tais parâmetros, chega-se em
$$
\alpha = 0, \quad \alpha' = \frac{\pi}{2}, \quad \beta = -\frac{\pi}{4}, \quad \beta' = \frac{\pi}{4},
\tag{10}
$$
o resulta em
$$
\left| \langle \phi_0 | \mathcal{C} | \phi_0 \rangle \right| = 2\sqrt{2} \approx 2.8284.
\tag{11}
$$
A equação (11) é conhecida como **limite de Tsirelson**, que representa o maior valor permitido pela teoria quântica, em contraste com o limite clássico $2$. Esse resultado evidencia a presença de correlações não-locais e constitui uma violação máxima da desigualdade de Bell-CHSH.

Na simulação realizada, chega-se ao seguinte gráfico:

```python
from PIL import Image
import matplotlib.pyplot as plt

# abrir a imagem
img = Image.open("Figure_2_3.png")

# mostrar a imagem
plt.imshow(img)
plt.axis("off")
plt.show()
```

É possível ver que o valor máximo de S atingido na simulação é justamente para o valor de Tsirelson ($S = 2\sqrt{2}$). A curva, $S(\theta)$, está presente tanto em regiões de regime não-local (quântido - zona vermelha), quanto de regime local (clássico - zona azul). Ademais, é notável que o intervalo $S \in [-2,\,2]$ representa **exatamente** a região clássica. 

O significado físico por trás desse resultado é importantíssima: como estamos maximizando a violação do primeiro estado de Bell, $S = 2\sqrt{2}$ significa que as correlações feitas são tão não-locais quanto a mecânica quântica permite. Em outras palavras, esse valor só pode ser atingido com estados maximamente emaranhados.

#### 3. $S = 4$

Como vimos nos exemplos anteriores, a formulação da desigualdade de Bell--CHSH é válida para um sistema bipartido $AB$. No entanto, existem generalizações da desigualdade de Bell--CHSH para sistemas multipartidos. Essas desigualdades generalizadas são conhecidas como desigualdades de Mermin [6]. 

Para um sistema tripartido ABC, o seguinte operador Hermitiniano é introduzido:

$$
\mathcal{M}_3 = \mathcal{M}_3^{\dagger}
= A' \otimes B \otimes C
+ A \otimes B' \otimes C
+ A \otimes B \otimes C'
- A' \otimes B' \otimes C'.
\tag{12}
$$
Usando as propriedades de norma e tendo em mente que cada comutador contribui com um fator $2$, obtêm-se
$$
\|\mathcal{M}_3^2\| = \|\mathcal{M}_3\|^2 \le 16.
\tag{13}
$$
Por conseguinte, para qualquer estado normalizado $|\psi\rangle$, $\langle \psi | \psi \rangle = 1$, tem-se que
$$
\left| \langle \psi | \mathcal{M}_3 | \psi \rangle \right| \le 4.
\tag{14}
$$
Com isso, afirma-se que o operador $\mathcal{M}_3$ é o operador de Mermin de ordem 3. Eles são os estados de Greenberger-Horne-Zeilinger [7], são, simplesmente, estados GHZ, e podem ser representados como

$$
|\psi_{\mathrm{GHZ}}\rangle
= \frac{1}{\sqrt{2}}
\left(
|+\rangle_A \otimes |+\rangle_B \otimes |+\rangle_C
- |-\rangle_A \otimes |-\rangle_B \otimes |-\rangle_C
\right).
\tag{15}
$$
Segundo Barnett et al. (2007), o estado GHZ, eq.(15), maximiza a inequação de inequação de Mermin. Ademais, o $S = 4$ é o máximo valor matematicamente possível do parâmetro CHSH, mas não pode ser alcançado pela mecânica quântica, dado que violaria a estrutura matemática do espaço de Hilbert e, com isso, permitiria correlações mais fortes que as quânticas. Isso representa um objeto teórico chamado PR-box (Popescu–Rohrlich box) [8] que é uma teoria hipotética mais não-local que a mecânica quântica, mas ainda sem comunicação superluminal. 

Em outras palavras, recorrendo à analogia realizada pelo Prolema 3, imagine que Alice e Bob sempre consigam correlacionar suas respostas perfeitamente, mesmo com escolhas aleatórias de medidas. Isso daria:
$$
AB = AB' = A'B = 1, \quad A'B' = -1
\tag{16}
$$
$$
S = 1 + 1 + 1 - (-1) = 4
\tag{17}
$$
O que não é possível em sistemas quânticos bipartidos com qubits reais (CHSH).
